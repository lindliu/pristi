```python
import matplotlib.pyplot as plt
import numpy as np 
import torch
import pickle
import pandas as pd
```

```python
def get_quantile(samples,q,dim=1):
    return np.quantile(samples,q,axis=dim)#.cpu().numpy()
```

```python
dataset = 'airquality' #choose 'healthcare' or 'airquality'
# dataset = 'healthcare'

datafolder_ours = 'pm25_outsample_20241226_102939_rearange with 2 layers'#'pretrained' # set the folder name
datafolder_pristi = 'pm25_outsample_20250130_131614'#'pretrained' # set the folder name
nsample = 100 # number of generated sample

path_ours = './PriSTI_m_5/save/'+datafolder_ours+'/generated_outputs_nsample' + str(nsample) + '.pk' 
path_pristi = './PriSTI-main/save/'+datafolder_pristi+'/generated_outputs_nsample' + str(nsample) + '.pk' 

with open(path_ours, 'rb') as f:
    samples,all_target,all_evalpoint,all_observed,all_observed_time,scaler,mean_scaler = pickle.load(f)

with open(path_pristi, 'rb') as f:
    samples_p,all_target_p,all_evalpoint_p,all_observed_p,all_observed_time_p,scaler_p,mean_scaler_p = pickle.load(f)

all_target_np = all_target.cpu().numpy()
all_evalpoint_np = all_evalpoint.cpu().numpy()
all_observed_np = all_observed.cpu().numpy()
all_given_np = all_observed_np - all_evalpoint_np

all_target_np_p = all_target_p.cpu().numpy()
all_evalpoint_np_p = all_evalpoint_p.cpu().numpy()
all_observed_np_p = all_observed_p.cpu().numpy()
all_given_np_p = all_observed_np_p - all_evalpoint_np_p


K = samples.shape[-1] #feature
L = samples.shape[-2] #time length

if dataset == 'airquality':
    path = './PriSTI-main/data/pm25/pm25_meanstd.pk'
    with open(path, 'rb') as f:
        train_mean,train_std = pickle.load(f)
    # train_std_cuda = torch.from_numpy(train_std).cuda()
    # train_mean_cuda = torch.from_numpy(train_mean).cuda()
    all_target_np=(all_target_np*train_std+train_mean)
    samples=(samples.cpu().numpy()*train_std+train_mean)

    all_target_np_p=(all_target_np_p*train_std+train_mean)
    samples_p=(samples_p.cpu().numpy()*train_std+train_mean)
    
elif dataset == 'healthcare':
    path = 'data/metr_la/metr_meanstd.pk'
    with open(path, 'rb') as f:
        train_mean,train_std = pickle.load(f)
    # train_std_cuda = torch.from_numpy(train_std).cuda()
    # train_mean_cuda = torch.from_numpy(train_mean).cuda()
    all_target_np=(all_target_np*train_std+train_mean)
    samples=(samples.cpu().numpy()*train_std+train_mean)
```

```python
qlist =[0.05,0.25,0.5,0.75,0.95]
quantiles_imp= []
for q in qlist:
    quantiles_imp.append(get_quantile(samples, q, dim=1)*(1-all_given_np) + all_target_np * all_given_np)

quantiles_imp_p= []
for q in qlist:
    quantiles_imp_p.append(get_quantile(samples_p, q, dim=1)*(1-all_given_np_p) + all_target_np_p * all_given_np_p)
```

```python
###airquality###
dataind = 17 #change to visualize a different sample

plt.rcParams["font.size"] = 16
fig, axes = plt.subplots(nrows=3, ncols=4,figsize=(20,10))
# fig.delaxes(axes[-1][-1])

for k_i,k in enumerate([0,1,22,23,    2,3,4,5,    32,33,34,35]):
    df = pd.DataFrame({"x":np.arange(0,L), "val":all_target_np[dataind,:,k], "y":all_evalpoint_np[dataind,:,k]})
    df = df[df.y != 0]
    df2 = pd.DataFrame({"x":np.arange(0,L), "val":all_target_np[dataind,:,k], "y":all_given_np[dataind,:,k]})
    df2 = df2[df2.y != 0]
    row = k_i // 4
    col = k_i % 4
    axes[row][col].plot(range(0,L), quantiles_imp[2][dataind,:,k], color = 'g',linestyle='solid',label='Ours')
    axes[row][col].fill_between(range(0,L), quantiles_imp[0][dataind,:,k],quantiles_imp[4][dataind,:,k],
                    color='g', alpha=0.3)
    axes[row][col].plot(df.x,df.val, color = 'b',marker = 'o', linestyle='None', label='target')
    axes[row][col].plot(df2.x,df2.val, color = 'r',marker = 'x', linestyle='None', label='condition')



    

    df = pd.DataFrame({"x":np.arange(0,L), "val":all_target_np_p[dataind,:,k], "y":all_evalpoint_np_p[dataind,:,k]})
    df = df[df.y != 0]
    df2 = pd.DataFrame({"x":np.arange(0,L), "val":all_target_np_p[dataind,:,k], "y":all_given_np_p[dataind,:,k]})
    df2 = df2[df2.y != 0]
    row = k_i // 4
    col = k_i % 4
    axes[row][col].plot(range(0,L), quantiles_imp_p[2][dataind,:,k], color = 'r',linestyle='solid',label='PriSTI')
    axes[row][col].fill_between(range(0,L), quantiles_imp_p[0][dataind,:,k],quantiles_imp_p[4][dataind,:,k],
                    color='r', alpha=0.3)
    axes[row][col].plot(df.x,df.val, color = 'b',marker = 'o', linestyle='None')
    axes[row][col].plot(df2.x,df2.val, color = 'r',marker = 'x', linestyle='None')


    if row==0 and col==0:
        axes[row][col].legend()
            
    if col == 0:
        plt.setp(axes[row, 0], ylabel='value')
    if row == -1:
        plt.setp(axes[-1, col], xlabel='time')

fig.tight_layout()

plt.savefig(f'compare_{dataind}.png',dpi=300)
```

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Simulated MMD values for 20 simulations per method
np.random.seed(42) # For reproducibility

methods = ["Vector AR", "LSTM", "Neural SDE", "E-MLP H_phi", "E-MLP K_phi"]
num_simulations = 20

# Simulated data (replace with actual values)
data = {
"Method": np.repeat(methods, num_simulations),
"MMD Score": np.concatenate([
np.random.normal(1.0, 0.1, num_simulations), # Vector AR
np.random.normal(0.5, 0.2, num_simulations), # LSTM
np.random.normal(0.3, 0.15, num_simulations), # Neural SDE
np.random.normal(0.02, 0.005, num_simulations), # E-MLP H_phi
np.random.normal(0.025, 0.006, num_simulations) # E-MLP K_phi
])
}
df = pd.DataFrame(data)

# Create violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x="Method", y="MMD Score", data=df, inner="quartile", palette="muted")
plt.xticks(rotation=45)
plt.title("Distribution of MMD Scores Across Methods")
plt.ylabel("MMD Score")
plt.xlabel("Method")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```
